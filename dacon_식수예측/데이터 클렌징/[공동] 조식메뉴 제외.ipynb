### 메뉴 데이터 추가분류
- 해당 데이터만 따로 리스트화해서 categorize()함수의 convert_word 자리에 공백으로 구분하여 두 개의 카테고리로 분류하기

```python
def categorize(data,category_list,convert_word):
    test = '|'.join(category_list)
    result = data.str.contains(test)
    result.fillna(False, inplace=True)
    data[result] = convert_word
```

## 빈도수 기반

```python
# 각 행에서 메뉴 데이터 count
# 해당 메뉴가 그 행에 있으면 1, 없으면 0 (몇 개 있느냐와 관계없이)
from tqdm import tqdm
menu_data_freq = menu_data.copy()
def count_menu():
  for i in tqdm(range(len(menu_data))):
      menu_data_freq.loc[i, '샐러드_count'] = 1 if True in menu_data.iloc[i].str.contains('샐러드').tolist() else 0
      menu_data_freq.loc[i, '빵_count'] = 1 if True in menu_data.iloc[i].str.contains('빵').tolist() else 0
      menu_data_freq.loc[i, '계란_count'] = 1 if True in menu_data.iloc[i].str.contains('계란').tolist() else 0
      menu_data_freq.loc[i, '주스_count'] = 1 if True in menu_data.iloc[i].str.contains('음료').tolist() else 0
      menu_data_freq.loc[i, '밥_count'] = 1 if True in menu_data.iloc[i].str.contains('밥').tolist() else 0
      menu_data_freq.loc[i, '면_count'] = 1 if True in menu_data.iloc[i].str.contains('면').tolist() else 0
      menu_data_freq.loc[i, '국_count'] = 1 if True in menu_data.iloc[i].str.contains('국').tolist() else 0
      menu_data_freq.loc[i, '생선류_count'] = 1 if True in menu_data.iloc[i].str.contains('생선류').tolist() else 0
      menu_data_freq.loc[i, '소고기_count'] = 1 if True in menu_data.iloc[i].str.contains('소고기').tolist() else 0
      menu_data_freq.loc[i, '닭고기_count'] = 1 if True in menu_data.iloc[i].str.contains('닭고기').tolist() else 0
      menu_data_freq.loc[i, '돼지고기_count'] = 1 if True in menu_data.iloc[i].str.contains('돼지고기').tolist() else 0
      menu_data_freq.loc[i, '오리고기_count'] = 1 if True in menu_data.iloc[i].str.contains('오리고기').tolist() else 0
      menu_data_freq.loc[i, '김치_count'] = 1 if True in menu_data.iloc[i].str.contains('김치').tolist() else 0
      menu_data_freq.loc[i, '과일_count'] = 1 if True in menu_data.iloc[i].str.contains('과일').tolist() else 0
      menu_data_freq.loc[i, '특별메뉴_count'] = 1 if True in menu_data.iloc[i].str.contains('특별메뉴').tolist() else 0
      menu_data_freq.loc[i, '기타_count'] = 1 if True in menu_data.iloc[i].str.contains('기타').tolist() else 0
    
  return menu_data_freq
```

```python
# 각 메뉴에 대한 count (전체 데이터 3762개)
menu_data_freq = count_menu()
menu_data_count = menu_data_freq.drop(['A','B','C','D','E','F','G','H','I'], axis = 1)
menu_data_count.sum()
```

```python
# 점수화하여 count 열 생성하는 함수
# [임시 점수화] 노션 참고

menu_data['menu_score'] = 0

def menu_score(df):
    for i in tqdm(range(len(df))):
        if True in menu_data.iloc[i].str.contains('특별메뉴').tolist():
            df.loc[i, 'menu_score'] += (5 * menu_data.iloc[i].str.contains('특별메뉴').sum())
        if True in menu_data.iloc[i].str.contains('돼지고기').tolist():
            df.loc[i, 'menu_score'] += (5 * menu_data.iloc[i].str.contains('돼지고기').sum())
        if True in menu_data.iloc[i].str.contains('생선류').tolist():
            df.loc[i, 'menu_score'] += (4 * menu_data.iloc[i].str.contains('생선류').sum())
        if True in menu_data.iloc[i].str.contains('소고기').tolist():
            df.loc[i, 'menu_score'] += (4 * menu_data.iloc[i].str.contains('소고기').sum())
        if True in menu_data.iloc[i].str.contains('과일').tolist():
            df.loc[i, 'menu_score'] += (3 * menu_data.iloc[i].str.contains('과일').sum())
        if True in menu_data.iloc[i].str.contains('샐러드').tolist():
            df.loc[i, 'menu_score'] += (3 * menu_data.iloc[i].str.contains('샐러드').sum())
        if True in menu_data.iloc[i].str.contains('닭고기').tolist():
            df.loc[i, 'menu_score'] += (3 * menu_data.iloc[i].str.contains('닭고기').sum())
        if True in menu_data.iloc[i].str.contains('면').tolist():
            df.loc[i, 'menu_score'] += (2 * menu_data.iloc[i].str.contains('면').sum())
        if True in menu_data.iloc[i].str.contains('오리고기').tolist():
            df.loc[i, 'menu_score'] += (1 * menu_data.iloc[i].str.contains('오리고기').sum())
    return df
```
