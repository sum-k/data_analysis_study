### 석식계 = 0 인 데이터 처리

```python
# 메뉴 데이터 생략없이 출력되도록 설정
pd.set_option('display.max_colwidth', -1)
```

```python
dinner_zero = data[data['석식계']==0]

# 석식메뉴 * 인 날
target_string = ['     *       ', '      *      ','    *        ']
dinner_star = data[data['석식메뉴'].map(lambda x: any(string in x for string in target_string))]
print(len(dinner_star))
dinner_star.head()
```

```python
def get_not_star_idx():
    dinner_star_idx = dinner_star.index.tolist()
    
    dinner_zero_list = dinner_zero.index.tolist()
    
    not_star_list = [idx for idx in dinner_zero_list if idx not in dinner_star_idx]
    
    return not_star_list
```

```python
# 특정 행사 or 메뉴 있는데 석식계 = 0 인 데이터
dinner_not_star = data.iloc[get_not_star_idx()]
print(len(dinner_not_star))
dinner_not_star.head(3)
```

```python
# * , 자기계발의 날, 가정의 날, 빈칸 -> 0 으로 통일 
# 석식 메뉴 있는 날 & 석식계 = 0은 정상 데이터로 생각하고 제거 -> 502, 410
dinner_not_star.drop([502,410], inplace=True)
```

```python
dinner_zero_df = pd.concat([dinner_star,dinner_not_star])
dinner_zero_idx = dinner_zero_df.index.tolist()
data.loc[dinner_zero_idx,'석식메뉴'] = '0'
```

### 메뉴 데이터 처리

```python
# 메뉴데이터만 추출
menu_features = ['조식메뉴','중식메뉴','석식메뉴']
menu_df = data[menu_features]
menu_df.head()
```

```python
# 처리하기 편하도록 하나의 리스트로 합치기 (0~1253 : 조식메뉴, 1254~2507 : 중식메뉴, 2508~ : 석식메뉴)
break_df = menu_df['조식메뉴']
lunch_df = menu_df['중식메뉴']
dinner_df = menu_df['석식메뉴']
menu_list = pd.concat([break_df,lunch_df,dinner_df],axis=0)
menu_list.reset_index(inplace=True,drop=True)
menu_list
```

```python
# 원산지 내용 제거하는 함수 작성
import re

def remove_origin():
    remove_list = []
    for x in menu_list:
        pattern = r'\([^)]*\)'
        x = str(x)
        text = re.sub(pattern=pattern, repl='', string= x)
        remove_list.append(text.split()) # 처리한 데이터 저장
    return remove_list
```

```python
# menu_list에 remove_origin() 함수 적용
remove_list = remove_origin()
remove_df = pd.DataFrame(remove_list)
remove_df.head()
```

```python
# 고추가루:중국산) <- 오류 데이터 찾기 / 6,7열에 2개씩 존재
remove_df.replace('고추가루:중국산)',None, inplace=True)
remove_list = remove_df.values.tolist()
```

```python
# 소스 데이터 제거 테스트
test_dt = remove_list[2290]
test_dt
```

```python
pat = r'\*.*$' # * 이후 오는 단어 제거 

for x in test_dt:
  if x :
    text = re.sub(pattern=pat, repl='', string = x)
    print(text)
```

```python
# *로 시작하는 양념/소스 데이터 제거하는 함수
def remove_sauce(data):
    remove_lt = []
    for x in data:
      if x : # None 인 데이터 제외하고 처리
        pattern = r'\*.*$'
        text = re.sub(pattern=pattern, repl='', string= x)
        remove_lt.append(text) # 처리한 데이터 저장
    return remove_lt
```

```python
remove_sauce(remove_df.loc[2290])
```

```python
# remove_df 행별로 remove_sauce 함수 적용
remove_df = remove_df.apply(remove_sauce, axis=1)
```
